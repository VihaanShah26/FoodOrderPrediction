# <b>Approach </b>
### Model Selection, Training and Testing 
I have used the DecisionTreeClassifier model. I have done so for the following reasons: 
- Easily Interpretable: Decision Trees provide a clear human-understandable representation of the decision making process. Each branch or leaf node corresponds to a specific condition established as a result of the features which makes it easy to understand why a particular prediction was made. 
- Non-Linearity: Decision trees capture complex relationships between multiple features and their overall impact on the target. Since there are 4 factors that are being considered in analysing their impact on the order placed, a simple linear model cannot be used to incorporate all the details provided. Decision trees provide a simple way of doing so. 
- Easy potential for increased accuracy: Decision Trees serve as the foundational blocks for ensemble methods such as RandomForestClassifiers and Gradient Boosters which usually yield a very high accuracy. 

Limitation: The Decision Tree Classifier is susceptible to overfitting and as a result it does not perform as accurately as more advanced algorithms such as ensemble methods or neural netowrks. 

I have split the data into a training set which is used to train the model and a testing set which the model performs predictions on and then the predictions can be compared to the traget value to check accuracy. The proportion used is 0.3 which means that 30% of the data is being used to train the model and 70% of the data is being used to test the model. Setting the random_state to be equal to a constant value ensures that each time the model is trained, the same split of data into training and testing data is used. This helps in performance comparisons. 

### Data Preperation
The data is read using the pandas library's read_csv function and is loaded into a pandas object. Since the data also contains categorical columns, the columns are converted into numerical labels using the LabelEncoder library. By doing this, it becomes easier to use different algorithms including ones that require numerical features. 


```python
# preparing the data 
def prepare_data(data): 
    # parameters: data - This is a pandas object. This is the csv file that was read using pandas 
    # returns: training and testing data , label encoders for categorical data 

    labels = {}
    columns = ["Year", "Major", "University", "Time", "Order"]

    # using labels to encode categorical columns so that numerical algorithms can be used on them 
    for col in columns: 
        labels[col] = LabelEncoder()
        data[col] = labels[col].fit_transform(data[col])
    
    # Splitting the data into features (X) and target (Y)
    X = data[["Year", "Major", "University", "Time"]]
    y = data["Order"]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    return X_train, X_test, y_train, y_test
```

```python
# training the model 
def train_model(X_train, y_train): 
    # selecting the model 
    model = DecisionTreeClassifier(max_depth=10, min_samples_split=10)
    # model = RandomForestClassifier(n_estimators=10, random_state=42)
    model.fit(X_train, y_train)

    # saving the model in a file using pickle so that it does not have to be trained each time 
    with open("trained_model.pkl", "wb") as trainedfile:
        pickle.dump(model, trainedfile)
```

```python
# function to call other functions and return a trained model which can be used to make predictions 
def get_model(X_train, y_train): 
    path_to_file = os.getcwd() + "/Dataset.csv"
    data = pd.read_csv(path_to_file)

    # check if the model has already been trained and pickled 
    path_to_trained_model = os.getcwd() + "/trained_model.pkl"
    if not os.path.exists(path_to_trained_model):
        # if the model has not been trained and pickled in the past, then train the model 
        train_model(X_train, y_train)
    
    # now the file trained_model.pkl is saved which is the pickled model which has been trained on the dataset 
    
    # loading the model 
    with open(path_to_trained_model, "rb") as trained_file: 
        model = pickle.load(trained_file)
        # the pickled model has now been loaded 
    
    return model 
```

```python
# function to use the model to make predictions
def make_predictions(model, X_test): 
    
    y_predictions = model.predict(X_test)
    
    return y_predictions
```

```python
# functions to test the accuracy of the predictions made using the model 
def test_accuracy(y_predictions, y_test): 
    
    # checking the accuracy of the predictions made 
    accuracy = accuracy_score(y_test, y_pred=y_predictions)
    print(f"Accuracy of the model = {accuracy}")
```

```python
# main function 
def main(): 
    # calling the function to prepare the data into machine readable data which can be used in the model 
    X_train, X_test, y_train, y_test = prepare_data(data)

    # calling the function to train the model/get the pickled model which can be used to make predictions 
    model = get_model(X_train, y_train)

    # calling the function to make predictions using the model 
    predictions = make_predictions(model, X_test)

    # testing the accuracy of the predictions made using the model 
    test_accuracy(y_predictions=predictions, y_test=y_test)
```

```python
if __name__ == "__main__": 
    main()
```
